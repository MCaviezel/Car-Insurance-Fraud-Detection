## Data manipulation to full dataset

```python
# No missing values --> no imputation needed
# Columns to drop
## incident_location all unique values --> no information (except without further geocoding)
## age since highly correlated with months_as_customer
## vechicle_claim since highly correlated with total_claim_amount
rows_to_delete = ['incident_location', 'age', 'vehicle_claim']
df.drop(rows_to_delete, axis = 1, inplace = True)
df.head()
```

```python
# Feature Creation from Date Variables
## Policiy bind date
df['policy_bind_date_day'] = df['policy_bind_date'].dt.day
df['policy_bind_date_month'] = df['policy_bind_date'].dt.month
df['policy_bind_date_dayofweek'] = df['policy_bind_date'].dt.dayofweek
df['policy_bind_date_dayofyear'] = df['policy_bind_date'].dt.dayofyear

## Incident date
df['incident_date_day'] = df['incident_date'].dt.day
df['incident_date_month'] = df['incident_date'].dt.month
df['incident_date_dayofweek'] = df['incident_date'].dt.dayofweek
df['incident_date_dayofyear'] = df['incident_date'].dt.dayofyear

## Delta in days between Policy bind date and Incident date
df['policy_to_incident'] = (df['incident_date']-df['policy_bind_date']).dt.days

## Delete two date columns
rows_to_delete = ['policy_bind_date', 'incident_date']
df.drop(rows_to_delete, axis = 1, inplace = True)
```

```python
# Round float to int
df['policy_annual_premium']=df['policy_annual_premium'].round(decimals=0).astype('int64')
```
